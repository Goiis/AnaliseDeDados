```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2,6,8]
y = [3,4,6]

x2 = [12,16,18]
y2 = [13,14,16]

#plt.bar chama o comando de barras, color muda a cor
plt.bar(x, y,label="Bar1",color="m")
plt.bar(x2, y2,label="Bar2", color="r")

plt.title("Gráficos \n Matheus Gois")
#plt.xlabel"x" define os nomes nas laterais
plt.xlabel("x")
plt.ylabel("y")
#insere a legenda
plt.legend()

#chama o gráfico
plt.show()
```

```python
population_ages = [17,17,17,10,20,20,5]

#ids = [x for x in range(len(population_ages))]

bins = [2,4,6,7,8,10,12,14,15,18,20,22,24,26]

plt.hist(population_ages,bins, histtype="bar", rwidth=0.8)

plt.title("Grafico de barras")
plt.xlabel("idades")
plt.ylabel("número de pessoas")

plt.show()
```

```python
x = [2,6,8]
y = [3,4,6]

plt.scatter(x,y,label="skitscat", color="k", marker="*", s=1000)
#s=100 define o tamanho
#marker o tipo do ponto, 
#para mais https://matplotlib.org/api/markers_api.html


plt.title("Gráficos \n Matheus Gois")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
dias = [1,2,3,4,5]

dormir = [7,8,6,11,7]
comer = [2,3,4,3,2]
trabalhar = [7,8,7,2,2]
jogar = [8,5,7,8,13]

#define legenda 
plt.plot([],[],color="m",label="dormir",linewidth=5)
plt.plot([],[],color="c",label="comer" ,linewidth=5)
plt.plot([],[],color="r",label="trabalhar" ,linewidth=5)
plt.plot([],[],color="k",label="jogar" ,linewidth=5)

#stackplot define o tipo de gráfico 
plt.stackplot(dias, dormir, comer, trabalhar, jogar, colors=["m","c","r","k"])

plt.title("Gráficos \n Matheus Gois")
plt.xlabel("x")
plt.ylabel("y")

#adiciona legenda
plt.legend()

plt.show()
```

```python
#define os podaços que representam cada parte
pedaços = [7,2,2,13]

#definindo o nome atribuido as cores na legenda
atividades = ["dormir","comer","trabalhar","jogar"]

#definindo a variavel das cores 
cores = ["c", "m", "r", "g"]

#startangle difine o ângulo de inicio do gráfico
#explode define um pedaço saindo do gráfico
#autopct acresenta a porcentagem no gráfico

plt.pie(pedaços, 
        labels=atividades, 
        colors=cores, 
        startangle= 90, 
        shadow=True, 
        explode=(0,0.1,0,0),
        autopct="%1.1f%%")

plt.title("Gráficos \n Matheus Gois")
plt.xlabel("x")
plt.ylabel("y")

#adiciona legenda
plt.legend()

plt.show()
```
