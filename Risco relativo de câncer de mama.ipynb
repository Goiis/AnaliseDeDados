# Risco relativo de câncer de mama

```python
x=[2.5,8.5,15.5,26.5] #Injestão de vinho
y=[1.00,1.08,1.15,1.22] #Risco

x1=[2.0,7.0,13.0,24.0] #Injestão de destilado
y1=[0.96,1.06,1.11,1.20] #Risco

plt.scatter(x,y, color="green", label="Vinho")
plt.scatter(x1,y1,label="Destilados")

plt.legend(loc="upper left")

plt.xlabel("Ingestão")
plt.ylabel("Risco")

plt.title("Risco relativo de câncer de mama")

plt.show()
```
