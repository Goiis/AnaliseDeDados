# Gráfico dos usuários do Facebook, Twitter e Linkedln:

##  Gráfico do Facebook

```python
import numpy as np
import  matplotlib.pyplot as plt
%matplotlib inline
```

```python
y  = [0, 10,0,14,0,19,0,17,0,17,0,13,0,10,0]
x= [ "",  "13 a 17","","18 a 24","","25 a 34","","35 a 44","","45 a 54","","55 a 64","","         acima de 65" ,"" ]

barWidth = 0.9
bar_color = '#0E4D92'
plt.xlabel('Idade')
plt.ylabel('Porcentagem(%)')
plt.title('Facebook')
plt.bar( np.arange( len( y ) ), y, barWidth,color=bar_color )
plt.xticks( np.arange( len( x ) ) + ( barWidth / 2 ), x )
plt.xlim( np.arange( len ( x ) ).min(), np.arange( len( x ) ).max() + barWidth)
plt.show()
```

# Gráfico de Setores: 

##  Gráfico do Facebook

```python
labels = "13 a 17","18 a 24","25 a 34","35 a 44","45 a 54","55 a 64","65 diante (10%)"
sizes = [10,14,19,17,17,13,10]
explode = (0, 0.3, 0, 0 ,0 ,0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%2.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
ax1.set(title='      FaceBook')

plt.show()
```
